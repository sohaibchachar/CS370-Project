Using the [CIFAR-10 dataset](https://www.cs.toronto.edu/~kriz/cifar.html), train a ResNet-50 based CNN on the classification task of $K=9$ classes (filter out the class `ship`) and create the following visualizations for first, middle and last blocks of ResNet-50. You are free to select a class to showcase such visualizations.

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        data_dict = pickle.load(fo, encoding='bytes')
    return data_dict

# Load the data batches
data_batches = []
for i in range(1, 6):
    batch = unpickle(f"cifar-10-batches-py/data_batch_{i}")
    data_batches.append(batch)

# Combine the data batches
train_images = np.vstack([batch[b'data'] for batch in data_batches])
train_labels = np.hstack([batch[b'labels'] for batch in data_batches])
mask = train_labels != 8

#import numpy as np
#indices = np.where(train_labels != 8)[0]
#filtered_train_images = train_images[indices]
#filtered_train_labels = train_labels[indices]




train_images = train_images[mask]
train_labels = train_labels[mask]
train_images = train_images.reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)
train_labels[train_labels > 8] -= 1
train_labels = to_categorical(train_labels, num_classes=9)
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D

base_model = ResNet50(weights=None, include_top=False, input_shape=(32, 32, 3))
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(9, activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=predictions)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=10, batch_size=32)


```

```python
model.save('Model1.h5')
```
